In this project, I was tasked with finding out whether hosting the olympics impacted the numbver of medals won by a country in that particular country and whther it affected the number of medals won and active participation in other olympics. I analyzed this data for countries in the summer and winter olympics to see what particular trend occued. I made a bar graph for each host country and indicated with an astrisck the year they hosted the olympics, and and exclamation point when they led with the most amount of medals won. I then graphed graphed 4-5 other olympics before and after the host year to see whether there was an increase or decline in medals won and participation. From this analysis, I was able to find that the years a country did not win that many medals, they didn't have that many participants. In addition, I also found that for most countries, they won the most medals when they were the host country and years after they were the host, they did,t have that much active participation and weren't winning any medals. For these countries, hosting the olympics was the highlight of fame and reputation for their country and has that small window to show for it. For other countries, such as the US, there weren't wasn't really any serious change in the number of medals won apart from when they were the host country. They were still able to win many medals and lead the olympics in those specific years. This could be explained by the rich economy and world power the US possesses and the large population it contains to always be a dominant force in the olympics. However, there where also countries, such as Canada in the winter olympics,that were near their lowest in the number of medals won when tehy hosted the olympics and had an increase when they weren't hosting the olympics.
